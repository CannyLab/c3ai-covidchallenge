```python
import math
counties_data = pd.DataFrame()
for i in range(math.ceil(len(countyList) / 10)):
    county_data = metr.fetch(countyList[i * 10 :min(len(countyList), i * 10 + 10)], ["JHU_ConfirmedCases"])
    for column in list(county_data.columns):
        if column.endswith("JHU_ConfirmedCases.data"):
            county_data = county_data.rename(columns={column: column.replace(".JHU_ConfirmedCases.data", "")})
        else:
            del(county_data[column])
    counties_data = pd.concat([counties_data, county_data], axis = 1)
counties_data = counties_data.T
```

```python
counties = c3aidatalake.fetch(
    "outbreaklocation",
    {
        "spec" : {
            "filter" : "contains(id, 'UnitedStates') && locationType == 'county'"
        }
    },
    get_all = True
)
counties = counties.set_index("id")
counties_data.columns = ["JHU_ConfirmedCases"]
for column in list(counties.columns):
    if column != "populationCDS":
        del(counties[column])
counties
```

```python
counties_data_combined = pd.concat([counties, counties_data], axis = 1)
rates = counties_data_combined["JHU_ConfirmedCases"]/ counties_data_combined["populationCDS"]
counties_data_combined = pd.concat([counties_data_combined, rates], axis = 1)
counties_data_combined.columns = ["populationCDS", "JHU_ConfirmedCases", "infection rate"]
counties_sorted = counties_data_combined.sort_values(by=['infection rate'])
counties_sorted
```
